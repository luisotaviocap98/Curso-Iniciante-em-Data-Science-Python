```python
import pandas as pd
dadosENEM = pd.read_csv('microdados_enem2018/DADOS/MICRODADOS_ENEM_2018.csv',sep=';',encoding='ISO-8859-1',nrows=200) #lendo 
colunas = ['NU_INSCRICAO', 'SG_UF_RESIDENCIA' ,'NU_IDADE','IN_GESTANTE', 'TP_SEXO' ,'TP_COR_RACA', 'TP_NACIONALIDADE',
 'TP_ESCOLA', 'TP_ENSINO' ,'CO_MUNICIPIO_PROVA' ,'NO_MUNICIPIO_PROVA' ,'CO_UF_PROVA', 'SG_UF_PROVA' ,'TP_PRESENCA_CN',
 'TP_PRESENCA_CH' ,'TP_PRESENCA_LC', 'TP_PRESENCA_MT' ,'CO_PROVA_CN' ,'CO_PROVA_CH' ,'CO_PROVA_LC' ,'CO_PROVA_MT' ,'NU_NOTA_CN',
 'NU_NOTA_CH' ,'NU_NOTA_LC' ,'NU_NOTA_MT' ,'TP_STATUS_REDACAO' ,'NU_NOTA_COMP1' ,'NU_NOTA_COMP2' ,'NU_NOTA_COMP3' ,'NU_NOTA_COMP4'
 'NU_NOTA_COMP5' ,'NU_NOTA_REDACAO','Q001','Q002'] #selecionando algumas colunas
dados_selecionados = dadosENEM.filter(items=colunas)
dados_selecionados.head()
```

```python
colQuest = ['NU_INSCRICAO','NU_NOTA_MT','NU_NOTA_REDACAO','Q001','Q002'] #selecionando colunas para analise das respostas socioeconomicas
socioEco = dados_selecionados.filter(items=colQuest)
questDict = {'A':'Nunca estudou.'
,'B':'No completou a 4 srie/5 ano do Ensino Fundamental.'
,'C':'Completou a 4 srie/5 ano, mas no completou a 8 srie/9 ano do Ensino Fundamental.'
,'D':'Completou a 8 srie/9 ano do Ensino Fundamental, mas no completou o Ensino Mdio.'
,'E':'Completou o Ensino Mdio, mas no completou a Faculdade.'
,'F':'Completou a Faculdade, mas no completou a Ps-graduao.'
,'G':'Completou a Ps-graduao.'
,'H':'No sei.'} #traducao das opcoes do questionario
```

```python
socioEco['SG_UF_RESIDENCIA'] = dados_selecionados.SG_UF_RESIDENCIA

# fg,ax = plt.subplots(figsize=(20,10)) 
# plt.suptitle('nota - escolaridade - estado')#titulo na figura

axes = socioEco.filter(items = ['SG_UF_RESIDENCIA','Q002','NU_NOTA_REDACAO']).where(socioEco.Q002 != 'H').groupby(['Q002','SG_UF_RESIDENCIA']).mean().sort_values(by='NU_NOTA_REDACAO',ascending=False).unstack().plot(figsize=(20,10)) #transformando isso tudo em um axes do matplot
axes.set_title("nota - escolaridade - estado") #colocando titulo no axes
# axes.legend(bbox_to_anchor=(1,1),loc=0,borderaxespad=0) #loc eh qual ponto da legenda entra em contato com o ponto ancora


import re
handle, label = axes.get_legend_handles_labels() 
newlabel = [re.search(',\s(.+?)\)',l).group(1) for l in label] #filtrando com regex apenas a sigla do estado
axes.legend(newlabel,bbox_to_anchor=(1,1),loc=0,borderaxespad=0) #bbox boundigbox eh a caixa que engloba a legenda

text=''
for i,j in questDict.items():
    text += '{k} : {v}\n'.format(k=i,v=j) #montando conjunto de frases com as traducoes
axes.text(0.02,0.6,text,transform=axes.transAxes,fontsize=10, color='white',bbox={'boxstyle':'round','facecolor':'black','alpha':0.8}) #colocando um texto no grafico com as traducoes
```

```python
colun = [ 'SG_UF_RESIDENCIA' ,'CO_MUNICIPIO_RESIDENCIA','NO_MUNICIPIO_PROVA' , 'SG_UF_PROVA' ,'NU_NOTA_CN', 'NU_NOTA_CH' ,'NU_NOTA_LC' ,'NU_NOTA_MT' ,'NU_NOTA_COMP1' ,'NU_NOTA_COMP2' ,'NU_NOTA_COMP3' ,'NU_NOTA_COMP4'
 'NU_NOTA_COMP5' ,'NU_NOTA_REDACAO'] #selecionando algumas colunas
novo_dado = dadosENEM.filter(items=colun)

for item in novo_dado.columns.values:
    novo_dado[item] = novo_dado[item].fillna(0) #substituindo os NaN

media = novo_dado.groupby(['SG_UF_RESIDENCIA','CO_MUNICIPIO_RESIDENCIA','NO_MUNICIPIO_PROVA'],as_index=False).mean() #as_index tira do negrito os agrupamentos
media #mostrando media de cada materia, agrupado por estado e municipio
```
